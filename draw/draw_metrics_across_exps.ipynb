```python
import json
import pandas as pd
import numpy as np
import pandas as pd
import json
import matplotlib.pyplot as plt

LAYERS_KEY = 'embedding_layer'


def jsonline2df(file_path):
    with open(file_path, 'r') as f:
        lines = f.readlines()
        datas = []
        columns = []
        for line in lines:
            data = json.loads(line)
            datas.append(data)
            columns = list(data.keys())
        df = pd.DataFrame(datas, columns=columns)
        return df

def plot_metrics_across_models(exps, metrics_key='orderness', matrics_name='Orderness'):
    plt.figure(figsize=(10, 6))
    for exp in exps:
        linstyle = exp.get('linstyle', '--')
        df = jsonline2df(exp['exp_path'])
        X = df[LAYERS_KEY]
        Y = df[metrics_key]
        plt.scatter(X, Y, label=exp['exp_name'], marker='o')
        plt.plot(X, Y, linestyle=linstyle)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.title("Orderliness across layers", fontsize=24)
    plt.xlabel("Layers", fontsize=24)
    plt.ylabel(matrics_name, fontsize=24)
    plt.yticks(fontsize=14)
    plt.xticks(fontsize=14)
    plt.legend(fontsize=14)
    plt.show()

        
        
```

# Analysis 5.1

```python
LAST_TOKEN_RESULTS_ROOT = f"outputs_last/results/"
METRICS_ACROSS_EXPS = [
    {
        'exp_name': 'Context-free number',
        'exp_path': f"{LAST_TOKEN_RESULTS_ROOT}/llama3.1-8b.nocontext_single.jsonl",
    },
    {
        'exp_name': 'cardNumber',
        'exp_path': f"{LAST_TOKEN_RESULTS_ROOT}/llama3.1-8b.cardNumber_singleDIY.jsonl",
        'is_averaged': True,
    },
    {
        'exp_name': 'meditation',
        'exp_path': f"{LAST_TOKEN_RESULTS_ROOT}/llama3.1-8b.meditation_single.jsonl",
        'is_averaged': True,
    },
    {
        'exp_name': 'focusdifference',
        'exp_path': f"{LAST_TOKEN_RESULTS_ROOT}/llama3.1-8b.focusdifference_single.jsonl",
        'is_averaged': True,
    },
    {
        'exp_name': 'forgetdifference',
        'exp_path': f"{LAST_TOKEN_RESULTS_ROOT}/llama3.1-8b.forgetdifference_single.jsonl",
        'is_averaged': True,
    },
    {
        'exp_name': 'shareMoney',
        'exp_path': f"{LAST_TOKEN_RESULTS_ROOT}/llama3.1-8b.shareMoney_singleDIY.jsonl",
        "linstyle": '-',
        'is_averaged': True,
    },
    {
        'exp_name': '1 + [%d]',
        'exp_path': f"{LAST_TOKEN_RESULTS_ROOT}/llama3.1-8b.additionLastNum_singleDIY.jsonl",
        "linstyle": '-',
        'is_averaged': True,
    },
    {
        'exp_name': 'totalDonation',
        'exp_path': f"{LAST_TOKEN_RESULTS_ROOT}/llama3.1-8b.totalDonation_3_singleDIY.jsonl",
        "linstyle": '-',
        'is_averaged': True,
    },
    {
        'exp_name': 'leftApple',
        'exp_path': f"{LAST_TOKEN_RESULTS_ROOT}/llama3.1-8b.leftApple_singleDIY.jsonl",
        "linstyle": '-',
        'is_averaged': True,
    },

]


def plot_average_trend(exps, metrics_key='orderness', matrics_name='Orderness'):
    plt.figure(figsize=(10, 6))
    Ys = []
    for exp in exps:
        df = jsonline2df(exp['exp_path'])
        X = df[LAYERS_KEY]
        Y = df[metrics_key]
        # plot non-averaged data
        if not exp.get('is_averaged', False):
            linstyle = exp.get('linstyle', '--')
            plt.scatter(X, Y, label=exp['exp_name'], marker='o')
            plt.plot(X, Y, linestyle=linstyle)
        else:
            Ys.append(Y)
    # plot averaged data
    Ys = np.array(Ys)
    Y_std = np.std(Ys, axis=0)
    Y_mean = np.mean(Ys, axis=0)
    plt.errorbar(X, Y_mean, yerr=Y_std, label='Context-based number, on average', marker='o', linestyle='--')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.title("Orderliness across layers", fontsize=24)
    plt.xlabel("Layers", fontsize=24)
    plt.ylabel(matrics_name, fontsize=24)
    plt.yticks(fontsize=14)
    plt.xticks(fontsize=14)
    plt.legend(fontsize=16)
    plt.savefig(f'./plot/first_descent.pdf', bbox_inches='tight', format="pdf", dpi=600)
    plt.show()

plot_average_trend(METRICS_ACROSS_EXPS)
```
